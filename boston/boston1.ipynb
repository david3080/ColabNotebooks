ボストン郊外地域の項目と住宅価値のデータを読み込みます。

```
import pandas as pd # ファイルを読み込むためのライブラリを読み込んで、プログラムからpdという名前で使えるよう宣言します。

house = pd.read_csv("https://raw.githubusercontent.com/david3080/ColabNotebooks/master/boston/BostonHousing.csv")
house
```

読み込んだデータから部屋数と住宅価値を配列として抜き出して、それらの関係を２次元グラフとして表します。

```
room = house['部屋数'].values
price = house['住宅価値'].values

import matplotlib.pyplot as plt # グラフを描画するためのライブラリを読み込んで、プログラムからpltという名前で使えるように宣言します。
plt.figure(figsize=(10, 5)) # グラフのサイズを設定します。
plt.scatter(room, price, c='black', s=3) # scatter（スキャター）関数で散布図を描きます（点の色（color）を黒、サイズを3ポイントに設定）。
plt.xlabel('ROOM', fontsize=14)
plt.ylabel('PRICE', fontsize=14)
plt.grid()
plt.show()
```

1行目が要素1で、2行目の要素が部屋数からなる2行506列の行列を作成します。

```
import numpy as np # 数値計算用のライブラリをnpとして読み込む

# 部屋数を表す一次元配列(room.values)に0列目を追加して、0列目（axis=1）に1.0をセットします
x = np.insert(room[:,np.newaxis], 0, 1.0, axis=1)
print(x)

# 行列数を表示します
print(x.shape)

# 行数をカウントします
M = x.shape[0]
print(M)
```

単回帰分析を行います。

```
M = x.shape[0] # 学習用のサンプルデータ数
alpha = 0.01 # 学習率
w = np.ones(2) # wの初期値。[1,1]のベクトル
yt = price # 
history = np.zeros((0,2)) # 損失数の記録

for k in range(100001):
    yp = x @ w # 学習データとwの内積をとって予測値を求める
    yd = yp - yt # 予測値と結果の差分
    w = w - alpha * (yd @ x)/M
    if ( k % 10000 == 0):
        loss = np.mean(yd ** 2) / 2 # 損失数
        history = np.vstack((history, np.array([k, loss]))) # 繰り返し数(k)と損失数(loss)の配列を追加
        print( "繰り返し数 = %d  損失数 = %f" % (k, loss))
    
print(w)
```

グラフ表示します。

```
x1=np.arange(room.min(), room.max(), 0.1)
y1=w[0] + w[1]*x1

plt.figure(figsize=(10, 5))
plt.scatter(x[:,1], yt, c='black', s=3)
plt.xlabel('ROOM', fontsize=14)
plt.ylabel('PRICE', fontsize=14)
plt.plot(x1, y1, color = 'red')
plt.grid()
plt.show()
```

学習曲線をグラフ表示します。

```
plt.figure(figsize=(10, 5))
plt.plot(history[:,0], history[:,1])
plt.grid()
plt.show()
```
