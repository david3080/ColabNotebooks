ボストン郊外地域の項目と住宅価値のデータを読み込みます。

```
import pandas as pd # ファイルを読み込むためのライブラリを読み込んで、プログラムからpdという名前で使えるよう宣言します。

house = pd.read_csv("https://raw.githubusercontent.com/david3080/ColabNotebooks/master/boston/BostonHousing.csv")
house
```

相関をヒートマップ化する

```
df_corr = house.corr()

import matplotlib.pyplot as plt
plt.figure(figsize=(20, 15))
import seaborn as sns
#sns.set(font='IPAGothic')
#sns.set(font='AppleGothic')
sns.heatmap(df_corr, annot=True, fmt='g', cmap='Blues')
```

住宅価格は部屋数と強いプラスの相関を持ち、低所得者率と強いマイナスの相関を持つことがわかります。
