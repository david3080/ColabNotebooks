```
import pandas as pd # ファイルを読み込むためのライブラリ
house = pd.read_csv("https://raw.githubusercontent.com/david3080/ColabNotebooks/master/boston/BostonHousing.csv")
house
```

低所得率と住宅価値の関係を単回帰分析します。

```
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
low = house[['低所得者率']].values
price = house['住宅価値'].values

lr.fit(low, price)
print('低所得率と住宅価値の関係：')
print('w0 = ', lr.intercept_)
print('w1 = ', lr.coef_[0])
```

低所得率と住宅価値の関係を単回帰分析した結果をグラフ化します。

```
import matplotlib.pyplot as plt # グラブ描画用のライブラリをpltとして読み込む

plt.scatter(low, price, color = 'blue')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(low, lr.predict(low), color = 'red') # 回帰直線をプロット

plt.xlabel('ROOM') # x軸のラベル
plt.ylabel('PRICE')  # y軸のラベル
plt.grid()                 # グリッド線を表示
plt.show()               # 図の表示
```
