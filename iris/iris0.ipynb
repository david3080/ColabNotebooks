あやめのがく片の長さと幅のデータを読み込み、ロジスティック回帰分析を行います。

```
import pandas as pd # ファイルを読み込むためのライブラリを読み込んで、プログラムからpdという名前で使えるよう宣言します。

iris = pd.read_csv("https://raw.githubusercontent.com/david3080/ColabNotebooks/master/iris/iris1.csv")
iris
```

あやめの品種ごとに点の色を変えて散布図を表示します。「品種」列の値が0の場合赤色、1の場合青色の点を描画します。

```
length = iris['がく片長'].values
width = iris['がく片幅'].values
species = iris['品種'].values

l0 = []
l1 = []
w0 = []
w1 = []
for i in range(len(length)):
  if species[i] == 0:
    l0.append(length[i])
    w0.append(width[i])
  else:
    l1.append(length[i])
    w1.append(width[i])

import matplotlib.pyplot as plt
plt.figure()
plt.scatter(l0, w0, c='red', s=3)
plt.scatter(l1, w1, c='blue', s=3)
plt.show()
```
